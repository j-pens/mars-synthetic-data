```python
import pandaset as ps

dataset = ps.DataSet('/zfs/penshorn/master_thesis/datasets/raw/PandaSet')

sequences = dataset.sequences()

sequences
```

```python
night_sequences = [ '057', '058', '059','063', '064', '065', '066', '067', '068', '069', '070', '071', '072', '073',
                       '074', '077', '078', '079', '149']
lighting_sequences = ['024','084', '085', '086', '088', '089', '090', '091', '092', '093', '094']

day_sequences = [seq for seq in sequences if seq not in night_sequences and seq not in lighting_sequences]

day_sequences
```

```python
assert len(day_sequences) + len(night_sequences) + len(lighting_sequences) == len(sequences) == 103
```

```python
import random

n_day_sequences = len(day_sequences)
n_day_test = n_day_sequences // 5

n_night_sequences = len(night_sequences)
n_night_test = n_night_sequences // 5

n_lighting_sequences = len(lighting_sequences)
n_lighting_test = n_lighting_sequences // 5


day_test_split = random.sample(day_sequences, n_day_test)
night_test_split = random.sample(night_sequences, n_night_test)
lighting_test_split = random.sample(lighting_sequences, n_lighting_test)

day_train_split = [seq for seq in day_sequences if seq not in day_test_split]
night_train_split = [seq for seq in night_sequences if seq not in night_test_split]
lighting_train_split = [seq for seq in lighting_sequences if seq not in lighting_test_split]

# n_day_train = len(day_train_split)
# n_day_val = n_day_train // 5

# n_night_train = len(night_train_split)
# n_night_val = n_night_train // 5

# n_lighting_train = len(lighting_train_split)
# n_lighting_val = n_lighting_train // 5

# day_val_split = random.sample(day_train_split, n_day_val)
# night_val_split = random.sample(night_train_split, n_night_val)
# lighting_val_split = random.sample(lighting_train_split, n_lighting_val)

# day_train_split = [seq for seq in day_train_split if seq not in day_val_split]
# night_train_split = [seq for seq in night_train_split if seq not in night_val_split]
# lighting_train_split = [seq for seq in lighting_train_split if seq not in lighting_val_split]

assert len(day_train_split) + len(day_test_split) == n_day_sequences
assert len(night_train_split) + len(night_test_split) == n_night_sequences
assert len(lighting_train_split) + len(lighting_test_split) == n_lighting_sequences

day_train_split, day_test_split
```
